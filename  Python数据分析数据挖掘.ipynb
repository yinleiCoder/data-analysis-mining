# 异常值处理
# 1. 利用四分数来确定异常值

- 四分位间距= 上四分位数(75%) - 下四分位数(25%)
- 数据下界 = 下四分位数 - k $ \times$ 四分位间距
- 数据上界 = 上四分位数 + k $ \times$ 四分位间距
- 异常值一般位于数据上下界之外。即如果大于数据上届or小于数据下界，则认为数据为异常值
- 一般情况下，k取1.5-3之间的数，k越小，正常值的范围越小，检测异常值的敏感性越高,k越大则正常值的范围越大。

```python
# 利用四分位间距检测异常值
for col in ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']:
    print('*'*50)
    print(col)
    # 求上四分位数
    q_75 = df[col].quantile(q=0.75)
    print('上四分位数', q_75)
        # 求下四分位数
    q_25 = df[col].quantile(q=0.25)
    print('下四分位数',q_25)
    # 求四分位间距
    d = q_75 - q_25
    print('四分位间距' ,d)
    # 求数据上界、下界
    data_top = q_75 + 1.5*d
    print('数据上界' ,data_top)
    data_bottom = q_25 -  1.5*d
    print('数据下界' ,data_bottom)
    # 查看异常值数量
    print('异常值个数', len(df[(df[col]>data_top) | (df[col]<data_bottom)]))
```
